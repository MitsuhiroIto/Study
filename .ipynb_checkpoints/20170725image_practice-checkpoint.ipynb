# 画像処理復習

---

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import json
import urllib2
from skimage.filters import sobel
%matplotlib inline
import tensorflow as tf
```

---

```python
murata = cv2.imread("murata.jpg")
print(type(murata))
print(murata.shape)
plt.imshow(murata)
plt.axis("off")
```

cv2.imreadで読み込むとBGRフォーマットなので、  
RGBに変換が必要（BlueとRedが入れ替わってる）  
なんかベイスターズっぽい

```python
murata_rgb = cv2.cvtColor(murata, cv2.COLOR_BGR2RGB)
print(murata_rgb.shape)
plt.imshow(murata_rgb)
```

```python
murata_gray = cv2.cvtColor(murata, cv2.COLOR_RGB2GRAY)
print(murata_gray.shape)
plt.imshow(murata_gray, cmap = "gray")
```

gray scale化

```python
plt.imshow(murata_rgb[:-70,160:-120])
print(murata_rgb[:-70,160:-120].shape)
```

拡大化

```python
_, threshold_img = cv2.threshold(murata_gray, 100, 200, cv2.THRESH_BINARY)
threshold_img = cv2.cvtColor(threshold_img, cv2.COLOR_GRAY2RGB)
plt.imshow(threshold_img)
```

```python
_, threshold_img = cv2.threshold(murata_gray, 100, 200, cv2.THRESH_BINARY_INV)
threshold_img = cv2.cvtColor(threshold_img, cv2.COLOR_GRAY2RGB)
plt.imshow(threshold_img)
```

閾値処理

```python
murata_bokashi = cv2.GaussianBlur(murata_rgb, (5,5), 0)
plt.imshow(murata_bokashi)
```

```python
murata_bokashi_bo = cv2.GaussianBlur(murata_gray, (5,5), 0)
plt.imshow(murata_bokashi_bo,cmap = "gray")
```

ぼかし

```python
_, threshold_img = cv2.threshold(murata_bokashi_bo, 100, 200, cv2.THRESH_BINARY_INV)
threshold_img = cv2.cvtColor(threshold_img, cv2.COLOR_GRAY2RGB)
plt.imshow(threshold_img)
```

からの閾値処理

```python
murata_f = np.fliplr(murata)
plt.imshow(murata_f)
```

画像フリップ

```python
murata_sobel = sobel(murata_gray)
plt.imshow(murata_sobel, cmap = "gray")
```
